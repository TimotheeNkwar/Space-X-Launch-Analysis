```python
# eda_visualization.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os

# Ensure output folder exists
os.makedirs("presentation", exist_ok=True)

# Load data
df = pd.read_csv("data/spacex_cleaned.csv")

# Summary statistics
print("Total launches:", len(df))
print(f"Success rate: {df['success'].mean() * 100:.2f}%")

# Launches per year
plt.figure(figsize=(10, 6))
launches_per_year = df.groupby("year").size()
sns.barplot(x=launches_per_year.index, y=launches_per_year.values)
plt.title("Launches per Year")
plt.xlabel("Year")
plt.ylabel("Number of Launches")
plt.savefig("presentation/launches_per_year.png")
plt.show()

# Success rate by rocket type
plt.figure(figsize=(8, 6))
success_by_rocket = df.groupby("rocket_name")["success"].mean() * 100
sns.barplot(x=success_by_rocket.index, y=success_by_rocket.values)
plt.title("Success Rate by Rocket Type")
plt.ylabel("Success Rate (%)")
plt.xticks(rotation=45)
plt.savefig("presentation/success_by_rocket.png")
plt.show()

# Launches by launchpad
plt.figure(figsize=(10, 6))
launches_by_launchpad = df.groupby("launchpad_name").size()
sns.barplot(x=launches_by_launchpad.values, y=launches_by_launchpad.index)
plt.title("Launches by Launchpad")
plt.xlabel("Number of Launches")
plt.savefig("presentation/launches_by_launchpad.png")
plt.show()

# Interactive scatter plot with Plotly
fig = px.scatter(
    df,
    x="year",
    y="success",
    color="rocket_name",
    title="Launch Success by Year and Rocket Type",
    labels={"success": "Success (True/False)", "year": "Year"}
)
fig.write_html("presentation/interactive_scatter.html")
```
